### **Online Business Sales 2017-2019**

This dataset contains data about online retail sales from an unknown business. This data uses two separate CSV files with different attributes related to the business' sales. The first CSV contains 6 features and 1775 records, and it includes variables such as online sales, profit, units sold, shipping, returns, and time of order. The second CSV contains 9 features and 36 records, and it includes variables such as month, year, total orders, gross sales, discounts, returns, net sales, shipping, and total sales.

The business had to reduce its workforce by 25 employees due to the economic downturn resulting from COVID-19.

```
# import necessary packages
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import pandas as pd
import numpy as np
```

```
# create first dataframe using first csv file
df = pd.read_csv("/content/business.retailsales.csv")
df.head()
```

```
# Find dimensions of the first csv file
df.shape
```

```
# create second dataframe using second csv file
df2 = pd.read_csv("/content/business.retailsales2.csv")
df2.head()
```

```
# Find dimensions of the second csv file
df2.shape
```

```
# Create pie chart

# data
products = 'Basket', 'Art & Sculpture', 'Jewelry', 'Kitchen'
sizes = [551, 337, 210, 161]
fig = plt.figure()
colors = ["lightgreen", "lightblue", "lightcoral", "plum", "lightorange"]
 
# pie chart attributes
plt.pie(sizes,
        labels = products,
        startangle=90,
        #colors=colors,  
        autopct='%.1f%%', 
        counterclock=True, 
        shadow=True)
plt.title('Most purchased products by category')
plt.axis('equal')
plt.tight_layout()

# show plot 
plt.show() 

# The product categories with the most purchases are baskets, art and sculpture supplies, jewelry, and kitchen supplies.
# Out of these categories, baskets have the most purchases.
```

```
# grouping data for line chart
a = df2.groupby('Month')['Total Orders'].sum()
lineX = list(a.keys())
lineY = list(a.values)
```

```
# grouping data for bar chart
b = df2.groupby('Month')['Net Sales'].sum()
barX = list(b.keys())
barY = list(b.values)
```

```
# Create bar chart

plt.bar(barX, barY)
plt.title("Net Sales are highest in November and December")
plt.xlabel("Month")
plt.ylabel("Net Sales")
plt.xticks(rotation = 45)
# Show Plot
plt.show()

# Sales are highest during November and December, in which sales range between $40,000 and $50,000.
# The rest of the year has sales that range between $15,000 and $30,000.
```

```
# Create scatter plot

plt.scatter(df2['Total Orders'], df2['Total Sales'], marker="o")
# Graph Properties
plt.title("Sales typically increase with order volume")
plt.ylabel("Total Sales")
plt.xlabel("Total Orders")

# Show Plot
plt.show()

# Most total orders lead to sales ranging from $4,000 to $18,000.
# Anything higher than that range may be an outlier or the result of special circumstances.
```

```
# Create histogram

ax = df2.hist(column='Total Orders', bins='auto', grid=False, rwidth=0.90)
ax = ax[0]
for x in ax:
    # Draw horizontal axis lines
    vals = x.get_yticks()
    x.set_title("Total Orders throughout 2017-2019 is skewed right")
    # Set x-axis label
    x.set_xlabel("Number of Total Orders")
    # Set y-axis label
    x.set_ylabel("Frequency")

# The business generally sees total order quantities below 100. 
```

```
# Create line chart
plt.xticks(rotation = 45)
plt.plot(lineX, lineY)
plt.title("Total orders change sporadically throughout the year")
plt.ylabel("Total Orders")
plt.xlabel("Month")
plt.show()

# Total order quantities fluctuate throughout the year, but they do exceptionally well during the holidays.
```

```
# Dashboard - arrange graphs in a 2D grid of columns and rows
%matplotlib inline
f, axs = plt.subplots(figsize=(18,10))
# spacing
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=.3, hspace=.4)

# scatter plot
plt.subplot(2, 3, 1)
plt.scatter(df2['Total Orders'], df2['Total Sales'], marker="o")
# Graph Properties
plt.title("Sales typically increase with order volume")
plt.ylabel("Total Sales")
plt.xlabel("Total Orders")

# histogram
plt.subplot(2, 3, 2)
ax = plt.hist(df2['Total Orders'], 
         histtype='stepfilled', 
         align='mid'
         )
plt.ylabel("Frequency")
plt.xlabel("Number of Total Orders")       
plt.title("Total Orders throughout 2017-2019 is skewed right")

# line
plt.subplot(2, 3, 3)
plt.xticks(rotation = 45)
plt.plot(lineX, lineY)
plt.title("Total orders change sporadically throughout the year")
plt.ylabel("Total Orders")
plt.xlabel("Month")

# bar
# plt.figure(figsize=(30,10)) 
plt.subplot(2, 3, 4)
plt.bar(barX, barY)
plt.xlabel("Month")
plt.ylabel("Net Sales")
plt.xticks(rotation = 45)
plt.title("5")
plt.title("Net Sales are highest in November and December")

# pie
plt.subplot(2, 3, 5)
plt.pie(sizes,
        labels = products,
        startangle=90,  
        autopct='%.1f%%',
        counterclock=True, 
        shadow=True)
plt.title('Most purchased products by category')
plt.axis('equal') 
plt.show()
```

#### **The 3-minute story**

The goal is to help this small business by understanding its products, analyzing its performance throughout the year, and finding ways to capitalize on its strengths while working on its weaknesses to obtain greater success as its plans for its future.

It is important to understand how this business makes money. It sells a variety of products, and it offers customers occasional discounts and allows for returns.
Fortunately, the company has customers who make a variety of orders. Thus, we see that total order amounts range from 50 to 350, and total sales increases with increasing order volume. Ideally, the more orders the business processes, the more they generate revenue; however, we can see that the business does better in producing about 50-100 orders on about 12-18 occasions. Total orders that exceed about 200 tend to be outliers. The company should focus on making orders that generate the most money per product. We also see that total orders change sporadically throughout the year. The business can improve on making consistent sales through continuously implementing strategic product pricing, placement, and promotion. The business has its peak performance in November and December generating over $40,000, as this is when they earn the highest net sales. They seem to do well during the holidays, but perhaps they need better discounts or offerings throughout the year. They do well in selling products such as baskets, art resources, jewelry, and kitchen supplies, but they do poorly in selling Easter craft, skincare, and furniture. The business might be able to capitalize on its niche specialty of providing accessories and creative supplies.
